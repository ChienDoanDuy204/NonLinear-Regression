```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
FILE_PATH = '/content/drive/MyDrive/AIO 2024/Module_4/Week 2/T7/Data/nonlinear_data.csv'
```

```python
df = pd.read_csv(FILE_PATH)
df.head()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
model = make_pipeline(PolynomialFeatures(degree=10), LinearRegression())
```

```python
model.fit(X,y)
```

```python
# Vẽ mô hình sau khi huấn luyện
plt.scatter(X, y, c = 'g',label = "data point train")
y_plot = model.predict(X)
plt.plot(X, y_plot, c = 'r', label = f"Polynomial model with degree = {9}")
plt.legend()
plt.grid()
plt.show()
```

```python
print(f"Lỗi của mô hình with degree = 10: ", mean_squared_error(y, model.predict(X)))
```

```python
model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
```

```python
model.fit(X,y)
```

```python
print(f"Lỗi của mô hình with degree = 2: ", mean_squared_error(y, model.predict(X)))
```
